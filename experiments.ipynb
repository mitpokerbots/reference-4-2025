# matplotlib

```python
data = []
value = 0
data.append(value)
for _ in range(1000):
    if np.random.random() < 0.5:
        value += 1
    else:
        value -= 1
    data.append(value)
data = np.array(data)
```

```python
data2 = []
value = 0
data2.append(value)
for _ in range(1000):
    if np.random.random() < 0.5:
        value += 1
    else:
        value -= 1
    data2.append(value)
data2 = np.array(data2)
```

```python
plt.plot(data, label = 'first data')
plt.plot(data2, label = 'other data')
plt.title('random walk')
plt.xlabel('steps')
plt.ylabel('value')
plt.legend()
plt.show()
```

```python
normal1 = np.random.normal(loc = -2, size=1000)
normal2 = np.random.normal(loc = 2, scale = 2, size=1000)
```

```python
plt.hist(normal1, bins=20, alpha=0.5)
plt.hist(normal2, bins=20, alpha=0.5)
plt.hist(normal1 + normal2, bins=20, alpha=0.5)
plt.show()
```

```python
plt.scatter(normal1, normal2+normal1, alpha=0.2)
plt.show()
```
